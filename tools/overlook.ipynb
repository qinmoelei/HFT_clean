```python
def find_position(actions):
    buy_positions = []
    sell_positions = []
    previous_action = 0
    for i in range(len(actions)):
        current_action = actions[i]
        if previous_action == 0 and current_action == 1:
            # print("all in")
            buy_positions.append(i)
        if previous_action == 1 and current_action == 0:
            # print("all out")
            sell_positions.append(i)
        previous_action = current_action
    holding_tuple = []
    for start, end in zip(buy_positions, sell_positions):
        holding_tuple.append((start, end))
    holding_lengths = []
    for (start, end) in holding_tuple:
        holding_lengths.append(end - start)

    return buy_positions, sell_positions, holding_tuple, holding_lengths

import numpy as np
action=np.load("/home/sunshuo/qml/Heta/result/result_group/env_trans_0back_time_length_1scale_factor_100/none_private_gamma_0.99_lr_0.001_64_nodes/test_result/action.npy")
buy_positions, sell_positions, holding_tuple, holding_lengths=find_position(action)
np.mean(holding_lengths)
```

```python
def max_drawdown(X):
    mdd = 0
    peak = X[0]
    for x in X:
        if x > peak: 
            peak = x
        dd = (peak - x) / peak
        if dd > mdd:
            mdd = dd
    return mdd  

return_rate_list=[]
sr_list=[]
mdd_list=[]
cr_list=[]
sor_list=[]
winning_rate_list=[]
for i in [12345,23451,34512,45123,51234]:
    returm_rate=np.load("/data/sunshuo/qml/HFT/result/standard_env_1/{}/return_rate.npy".format(i))
    return_sum=np.sum(returm_rate)
    sr=np.mean(returm_rate)/np.std(returm_rate)
    actual_reward=np.load("/data/sunshuo/qml/HFT/result/standard_env_1/{}/reward.npy".format(i))
    neg_ret_lst=[r for r in actual_reward if r <0]
    winning_rate_list.append(1-len(neg_ret_lst)/len(actual_reward))
    total_asset=[250]
    for reward in actual_reward:
        total_asset.append(total_asset[-1]+reward)
    mdd=max_drawdown(total_asset)
    cr = np.sum(returm_rate) / mdd
    sor = np.sum(actual_reward)/np.std(neg_ret_lst) 
    return_rate_list.append(return_sum)
    sr_list.append(sr)
    mdd_list.append(mdd)
    cr_list.append(cr)
    sor_list.append(sor)
print("mean return",np.mean(return_rate_list),"std_return",np.std(return_rate_list))
print("mean sr",np.mean(sr_list),"std_sr",np.std(sr_list))
print("mean mdd",np.mean(mdd_list),"std_mdd",np.std(mdd_list))
print("mean cr",np.mean(cr_list),"cr_return",np.std(cr_list))
print("mean sor",np.mean(sor_list),"sor_return",np.std(sor_list))
print("mean sor",np.mean(sor_list),"sor_return",np.std(sor_list))
print("mean wining",np.mean(winning_rate_list),"sor_return",np.std(winning_rate_list))



    
```
